```python
# Create a simple pandas series from a list
import pandas as pd

a=[1,3,4,5,6]
myvar=pd.Series(a)
print(myvar)
```

```python
#label
print(myvar[2])
```

```python
import pandas as pd

a= [1,2,3]
myvar = pd.Series(a,index=["x","y","z"])

print(myvar)
```

```python
import pandas as pd

calories = {
    "day 1": 420,
    "day 2": 390,
    "day 3": 500
}

myval= pd.Series(calories)
print(myval)
```
